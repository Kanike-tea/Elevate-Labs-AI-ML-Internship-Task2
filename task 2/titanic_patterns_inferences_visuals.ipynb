# Titanic Dataset: Patterns, Trends, Anomalies & Inferences

This notebook focuses on:
- Identifying patterns, trends, or anomalies
- Making feature-level inferences from visuals

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='whitegrid')

df = pd.read_csv('Titanic-Dataset.csv')
df.head()
```

## Sex vs Survival
- Females had a much higher survival rate than males.

```python
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title('Survival by Sex')
plt.show()
```

## Pclass vs Survival
- 1st class passengers had higher survival than 3rd class.

```python
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title('Survival by Passenger Class')
plt.show()
```

## Age Distribution by Survival
- Young children had better survival odds.

```python
sns.violinplot(data=df, x='Survived', y='Age', inner='quartile')
plt.title('Age Distribution by Survival')
plt.show()
```

## Fare Outliers
- A few very high fares are outliers.

```python
sns.boxplot(x=df['Fare'])
plt.title('Fare Distribution (Outliers Visible)')
plt.show()
```

## Embarked vs Survival
- Port C showed slightly higher survival rate.

```python
sns.countplot(data=df, x='Embarked', hue='Survived')
plt.title('Survival by Embarkation Port')
plt.show()
```

## Inferences Summary
- **Females** had a much better chance of survival.
- **1st class** passengers were more likely to survive.
- **Children** (especially under 10) had better survival odds.
- **High fares** were associated with survival but had **outliers**.
- **Embarked = C** passengers slightly more likely to survive.
